```python
# Importar paquetes
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Funcion que calcula la pdf para la distribuci√≥n normal
def normal_pdf(x, mu=0, sigma=1):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
```

Si queremos calcular $ e^2 $ lo podemos hacer con la funcion de numpy `np.exp(2)`

```python
# La funcion de scipy scipy.stats.norm.pdf() es equivalente
# ejemplo
print(normal_pdf(2, mu=0, sigma=1))
print(norm.pdf(2, loc=0, scale=1))
```

```python
x_axis = np.linspace(-8, 8, 1000)
y_values = []
for i in x_axis:
    y_values.append(normal_pdf(i, mu=0, sigma=1))
```

```python
np.sum(norm.pdf(x_axis, loc=0, scale=1))
```

```python
plt.figure()
plt.plot(x_axis, y_values)
```

```python
cumulative_y_values = []
for index in range(len(y_values)):
    cumulative = np.sum(y_values[:index])
    cumulative_y_values.append(cumulative)
# Normalizar
cumulative_y_values = cumulative_y_values / np.sum(cumulative)
```

```python
plt.figure()
plt.plot(x_axis, cumulative_y_values)
```

```python
y_values_scipy = norm.cdf(x_axis, loc=0, scale=1)
plt.figure()
plt.plot(x_axis, y_values_scipy)
```
